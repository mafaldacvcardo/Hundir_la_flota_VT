```python
#Importa bibliotecas que nos van a hacer falta: numpy para la tabla y random para elecciones aleatorias.
import numpy as np
import random

#Primero definimos todas las funciones que nos van a hacer falta.

#Define la función para crear el tablero.
#Por defecto, el tablero tendrá un tamaño de 10, pero si al utilizar la función le ponemos otro valor a "tamaño" nos creará un tablero de otro tamaño.
#Creamos un tablero con np.full. Como las dos variables son "tamaño", cogerá el valor que le metamos a la función y creará un tablero cuadrado. Rellenará las posiciones del tablero con lo que le digamos, en este caos, "_".

#IMPORTANTE: cogí el tablero que me creó ChatGPT porque me parecía más claro. Pero si tenéis que usar el que hicisteis en clase habría que cambiar eso.

#Esta función "crear_tablero" devuelve el "tablero" creado.
def crear_tablero(tamaño=10):
    tablero = np.full((tamaño, tamaño), "_")
    return tablero


#Define la función para colocar en el tablero los barcos que le digamos.
#El primer parámetro, "barco", será una lista con las posiciones/coordenadas que ocupe el barco.
#El segundo parámetro, "tablero", le dice a la función en qué tablero tiene que colocar el barco.
#La función "for" va pasando, de una en una, por todas las posiciones/coordenadas de la lista que compone el barco.
#Por ejemplo, si tienes un barco [(0,1), (0,2), (0,3)], irá a la posición (0,1) y sustituirá lo que había ("_") por "O".
#Luego pasará a la siguiente posición (0,2) y hará lo mismo. Seguirá así hasta que se acabe el barco.
#Así, nos habrá puesto "O" en las posiciones que está el barco y el barco ya estará situado en el tablero.
def colocar_barcos(barco, tablero):
    for (fila, columna) in barco:
        tablero[fila, columna] = "O"


#Define la función para disparar, que tendrá tres parámetros: casilla, tablero y barcos.
#El parámetro casilla se formará por los valores que metamos en cada turno (con input) para fila y columna.
#El tablero en el que queremos que dispare.
#Una lista con los barcos que hay en el tablero. Se utiliza para comprobar si el disparo golpea/toca/hunde algún barco.
#Mira si en la casilla (fila, columna) que le decimos hay algún barco (casilla con "O").
#Si en la casilla donde disparas hay algún barco, cambiará la "O" por la "X"; es decir, que hemos golpeado un barco con nuestro disparo.
#En tal caso, imprimirá "¡Tocado!".
#Con el "for" revisará cada barco, de uno en uno, de todos los barcos que hay en el tablero (en la lista "barcos").
#Aquí llama a la función barco_hundido (aparece y se explica después). Si esa función barco_hundido nos devuelve "True", quiere decir que el barco está hundido e imprimimos "¡Hundido!".
#Si el barco está hundido, lo quitamos de la lista de barcos con .remove y cortamos el ciclo con break.
#Y nos devuelve "tocado", así el programa sabe que ese barco está tocado.
#Con el "elif" lo que tenemos es la alternativa al primer "if". Es decir, si la casilla a la que disparamos no tiene barco "no tiene "O", sino que está vacía "_", cambiamos "_" por "A" (de agua) en esa posición donde hemos disparado e imprimimos "¡Agua!".
#Con el "else", si no tenemos "O" ni "_" en esa casilla, quiere decir que tenemos "X" o "A", así que ya hemos disparado ahí antes. Y con el return "repetido" se guarda la información de que ya has disparado ahí antes.
def disparar(casilla, tablero, barcos):
    fila, columna = casilla
    if tablero[fila, columna] == "O":
        tablero[fila, columna] = "X"
        print("¡Tocado!")
        for barco in barcos:
            if barco_hundido(barco, tablero):
                print("¡Hundido!")
                barcos.remove(barco)
                break
        return "tocado"
    elif tablero[fila, columna] == "_":
        tablero[fila, columna] = "*"
        print("¡Agua!")
        return "agua"
    else:
        print("¡Ya disparaste ahí antes!")
        return "repetido"


#Define la función para comprobar si el barco está solamente tocado o hundido (ha sido disparado en todas sus posiciones).
#Tendrá dos parámetros: la lista con las posiciones donde está el barco y el tablero donde se encuentra el barco.
#Con "for", lo que hacemos es pasar de una en una por cada posición de un barco.
#En cada posición del barco mira si es distinta a "X". Si alguna posición de un barco es distinta a X es que no le han disparado en todas sus posiciones, por lo que no estará hundido (False).
#En cambio, si la posición es igual a "X", comprueba la siguiente posición y así hasta comprobar todas las posiciones del barco.
# Si todas las posiciones son igual a X, devolverá True, lo que quiere decir que el barco está hundido.
def barco_hundido(barco, tablero):
    for f, c in barco:
        if tablero[f, c] != "X":
            return False
    return True


#Define la función para crear un barco. Utiliza dos parámetros para crearlo y asegurar que está dentro del tablero: eslora (quiere decir la longitud) y el tamaño del tablero.
#Primero le decimos al programa que decida aleatoriamente (random.choice) si el barco es horizontal o no.
#Si es horizontal, asignará a la fila un valor aleatoriamente con random.randint (será un número entero, por eso int). Le decimos que este número aleatorio tiene que estar entre el 0 y el tamaño-1, es decir, entre la posición 0 y la 9.
#Para elegir aleatoriamente en qué columna empieza y asegurar que el barco no se sale del tablero, utilizamos (0, tamaño_tablero - eslora).
#Es decir, nos da igual cómo de a la izquierda empiece (0), pero lo máximo a la derecha que puede empezar es el tamaño - la eslora.
#Por ejemplo, si queremos meter un barco con eslora 4, la columna en la que empiece tendrá que ser alguna entre 0 y 6 (10-4). Si empieza en 0 entre, obviamente, y si empieza en 6 también: barco en 6, 7, 8 y 9.
#Una vez tiene aleatoriamente la fila donde se situa el barco y la columna donde empieza, lo que hace es crear el barco.
#El barco será una lista con todas las posiciones donde se encuentra el barco.
#En este caso, como el barco es horizontal, se van añadiendo posiciones manteniendo la fila y cambiando la columna (añadiendo la "i").
#Con el "for" lo que hace es ir pasando por i=0, i=1, etc., hasta completar la eslora del barco. Por ejemplo, si la eslora es 3, empezará en i=0, luego pasará a i=1 y luego a i=2, porque le hemos dicho que "i" vaya cogiendo valores según range(3), es decir, de 0 a 2.
#Así, nos creará un barco de la eslora que le digamos (lo dice en las normas del juego, 3 barcos de eslora 2, 2 barcos de eslora 3 y 1 barco de eslora 4).
#Por ejemplo, si es un barco de eslora 3, si la fila de inicio aleatoria ha sido la 5 y la columna de inicio aleatoria ha sido la 2, tendremos el barco [(5,2), (5,3), (5,4)].
#Si al principio la función decide aleatoriamente que el barco de vertical, se hará lo mismo, pero cambiando fila por columna.
#Esta función se utilizará dentro de generar_barcos_aleatorios (se explica después).

def crear_barco(eslora, tamaño_tablero=10):
    horizontal = random.choice([True, False])
    if horizontal:
        fila = random.randint(0, tamaño_tablero - 1)
        columna_inicio = random.randint(0, tamaño_tablero - eslora)
        barco = [(fila, columna_inicio + i) for i in range(eslora)]
    else:
        columna = random.randint(0, tamaño_tablero - 1)
        fila_inicio = random.randint(0, tamaño_tablero - eslora)
        barco = [(fila_inicio + i, columna) for i in range(eslora)]
    return barco

#Define la función para generar barcos aleatorios. Esta función genera todos los barcos necesarios (si le damos la lista de esloras) y para ello utiliza la función de crear_barco que hemos visto arriba.
#Primero, creamos la lista vacía "barcos" que iremos rellenando: barcos = []
#Más adelante, en la función "jugar", veremos que le damos al programa la lista de esloras, que será [2, 2, 2, 3, 3, 4] (sacado de las normas del juego).
#Con "for", lo que hacemos es ir pasando por todas las esloras de la lista (2, luego 2, luego 2, luego 3, etc.), de los barcos que queremos crear. En nuestro caso, queremos crear seis barcos).
#Entonces, con el while, iniciamos un bucle que solamente se parará cuando se haya generado un barco correctamente, sin posiciones superpuestas encima de otro barco, posición ya ocupada.
#Creamos un barco con la función crear_barco que hemos visto más arriba. Como parámetros, cogerá la eslora y el tamaño del tablero.
#Tenemos que ver si el barco que crea es válido o no, es decir, que no se coloque encima de otro que ya esté colocado.
#Damos por bueno el barco que crea (True), pero con el "for f, c in barco" vamos a comprobar que de verdad es válido.
#Vamos pasando por las posiciones del barco f, c (iremos comprobando una a una) y comprobamos si esas posiciones están vacías ("_") en el tablero.
#Si esa posición es distinta a "_" quiere decir que hay ya algo, así que el barco pasaría a no ser válido (False).
#Si el barco no es válido (barco_valido = False), con el break salimos del "for f, c in barco" y volvemos al while de arriba, empezando otra vez el proceso de crear el barco con esa misma eslora.
#Volverá a entrar al "for" para comprobar si ese barco aleatorio se ha posicionado bien. Si se ha posicionado bien y no está superpuesto (es decir, barco_valido = True), con la función "colocar_barcos" (explicada arriba) colocará los barcos en el tablero y pondrá "O" en las posiciones donde esté el barco.
#También, con barcos.append(barco), irá añadiendo los barcos que va creando a la lista barcos que hemos creado vacía más arriba.
#Una vez hecho esto, con el break, corta el while y pasa a la siguiente eslora de la lista_esloras. Hará lo mismo para cada una de las esloras, hasta crear y posicionar bien todos los barcos.
#Al final, devuelve la lista de barcos, que tendrá todos los barcos creados con todas sus posiciones. Por ejemplo, si fuesen dos barcos, uno de dos y otro de eslora 3, podríamos tener:
#barcos = [[(3,6),(3,7)], [(0,0),(1,0),(2,0)]]. En cada while iría sumando los barcos a esta lista.
def generar_barcos_aleatorios(tablero, lista_esloras):
    barcos = []
    for eslora in lista_esloras:
        while True:
            barco = crear_barco(eslora, tamaño_tablero=10)
            barco_valido = True
            for f, c in barco:
                if tablero[f, c] != "_":
                    barco_valido = False
                    break
            if barco_valido:
                colocar_barcos(barco, tablero)
                barcos.append(barco)
                break      
    return barcos

#Definimos la función que utilizaremos para mostrar los tableros después de cada turno. Tiene dos parámetros: el tablero base que creamos y ocultar_barcos, que por defecto será False. Más adelante lo pondremos como True a la hora de mostrar el tablero de la máquina, donde nosotros disparamos, para que no veamos dónde tiene los barcos la máquina.
#Si ocultar_barco es False, se mostrarán los barcos "O"; si es True, sustituirá las "O" por "_", para que se muestren todas las casilla iguales.

def mostrar_tablero(tablero, ocultar_barcos=False):
    for fila in tablero:
        fila_mostrar = []
        for casilla in fila:
            if ocultar_barcos and casilla == "O":
                fila_mostrar.append("_")
            else:
                fila_mostrar.append(casilla)
        print(" ".join(fila_mostrar))


#Definimos la función para comprobar si todos los barcos están hundidos en un tablero.
#El parámetro de esta función será el tablero.
#Con el primer for, irá pasando por todas las filas.
#Dentro de este for está el segundo for, que le hace pasar por cada casilla de una fila. Es decir, estos dos for, uno dentro del otro, lo que hacen es que empiece por la primera fila y compruebe todas las casillas. Luego pasa a la segunda fila y comprueba todas las casillas, etc. Así, comprueba toda el tablero completo.
#Si alguna casilla es igual a "O", nos devolverá False: es decir, todos los barcos no están hundidos, todavía queda algún barco (por lo menos) sin hundir.
#Si ninguna casilla es igual a "O", es decir, si no encuentra ningún barco en el tablero, quiere decir que todos los barcos están hundidos y nos devolverá True.

def todos_hundidos(tablero):
    for fila in tablero:
        for casilla in fila:
            if casilla == "O":
                return False
    return True

#Definimos la función turno_jugador, que tiene dos parámetros: tablero_maquina y barcos_maquina.
#Importante: tendremos dos tableros en el juego. Uno será el nuestro (al que disparará la máquina y será totalmente visible para nosotros). El otro será el de la máquina, donde dispararemos nosotros, así que solamente serán visibles los * (agua), X (tocado), pero no las "O".
#El while True lo que hace es que genera un bucle que se mantiene hasta que hacemos un disparo válido.
#Primero pide que metamos la fila donde queremos disparar.
#Después nos pide que metamos la columna donde queremos disparar.
#Con el "if not" lo que hacemos es comprobar si lo que mete el usuario son números. Si no son números, imprime "Introduce números válidos" y te volverá a pedir que metas la fila y la columna.
#Si son números, pasa la fila y la columna a números enteros con "int".
#Si tanto la fila como la columna no están entre 0 y 9, el disparo es fallido, te imprime "Coordenadas fuera del tablero" y con el continue vuelves al principio y te pedirá otra vez la fila y la columna.
#Si la fila y la columna que metes están entre el 0 y el 9, creamos una variable "resultado" y llamamos a la función "disparar" (la hemos visto más arriba).
#Con esto lo que queremos es poder comparar la variable "resultado" con los return que teníamos en la función "disparar".
#Si disparamos a una posición donde ya habíamos disparado antes (return "repetido"), con continue volvemos arriba y nos permitirá volver a meter la fila y la columna donde queremos disparar.
#Si la variable resultado no es igual a "repetido", con el break cortamos el bucle while True y el turno pasa a la máquina.

def turno_jugador(tablero_maquina, barcos_maquina):
    while True:
        fila_input = input("Fila (0-9): ")
        columna_input = input("Columna (0-9): ")
        if not (fila_input.isdigit() and columna_input.isdigit()):
            print("Introduce números válidos (0-9).")
            continue
        fila = int(fila_input)
        columna = int(columna_input)
        if not (0 <= fila < 10 and 0 <= columna < 10):
            print("Casilla fuera del tablero.")
            continue
        resultado = disparar((fila, columna), tablero_maquina, barcos_maquina)
        if resultado == "repetido":
            continue
        break

#Definimos la función del turno de la máquina, que tendrá tres parámetros: tablero jugador (el mío), barcos_jugador (los míos) y disparos_previos, que comprueba que la máquina, con sus disparos random, no haya disparado a la misma casilla antes.
#Abrimos el bucle con while True que no se cortará hasta que la máquina no haga un disparo válido.
#Se elige la fila donde disparará la máquina de forma aleatoria con el random.randint, que será un número entero entre 0 y 9.
#Se elige la columna donde disparará la máquina de forma aleatoria con el random.randint, que será un número entero entre 0 y 9.
#Se coge el valor de fila y de columna y comprobamos si la máquina ya ha disparado a esa posición antes, comprobando si está en disparos_previos.
#Si no está en disparos_previos, añadimos este disparo a disparos_previos con .add para que se vaya actualizando.
#En resultado llamamos a la función "disparar" (explicada más arriba) y le decimos que dispare a esa fila y columna, en el tablero del jugador que contiene los barcos del jugador.
#Con el disparo, según vimos en la función disparar, dependiendo de qué haya en la casilla donde dispara la máquina pasará a "X" si había un barco ("O") o "A" si no había nada "_".
#Inprime a dónde ha disparado la máquina para que sea más visible y no tengamos que estar mirando y comparando nuestra table con la tabla del turno anterior.
#Con el break se corta el bucle while True y el turno pasa al jugador.

def turno_maquina(tablero_jugador, barcos_jugador, disparos_previos):
    while True:
        fila = random.randint(0, 9)
        columna = random.randint(0, 9)
        if (fila, columna) not in disparos_previos:
            disparos_previos.add((fila, columna))
            resultado = disparar((fila, columna), tablero_jugador, barcos_jugador)
            print(f"La máquina ha disparado a ({fila}, {columna})")
            print("--------------------------")
            break

#Define la función jugar.
#Crea tablero_jugador con la función crear_tablero(), que por defecto tiene tamaño 10 (la hemos visto más arriba).
#Crea tablero_maquina con la función crear_tablero(), que por defecto tiene tamaño 10 (la hemos visto más arriba).
#La lista de esloras. Lo que dice es que queremos 3 barcos de eslora 2, 2 barcos de eslora 3 y 1 barco de eslora 4.
#Generamos los barcos del jugador (barcos_jugador) con la función generar_barcos_aleatorios. La máquina disparará a estos barcos.
#Generamos los barcos la máquina (barcos_máquina) con la función generar_barcos_aleatorios. El jugador disparará a estos barcos.
#Creamos un set vacío disparos_previos, donde se irán añadiendo los disparos que vaya haciendo la máquina (lo hemos visto en la función turno_maquina).
#Con while True empieza el bucle. Terminará cuando algún jugador gane la partida.
#Imprime "tú tablero" y muestra tu tablero (llamando a la función tablero_jugador, definida y explicada más arriba) seguido; por defecto, si no se dice lo contrario, se muestran los barcos en el tablero.
#Imprime"tablero máquina" y muestra el tablero de la máquina (llamando a la función tablero_maquina, definida y explicada más arriba) con los barcos ocultos, porque es donde vamos a disparar y donde tenemos que adivinar dónde están situados los barcos.
#El primer turno es el nuestro, el del jugador (llama a la función turno_jugador, definida y explicada más arriba). Los dos parámetros son tablero_maquina y barcos_maquina, porque en nuestro turno queremos disparar al tablero de la máquina.
#Después de cada turno, con el if, llamamos a la función todos_hundidos, cuyo parámetro será, en este caso, el tablero_maquina, que es donde queremos comprobar si queda algún barco a flote.
#Si todos los barcos estuviesen hundidos, imprimiría el mensaje "¡Has ganado! ..." y cortará el bucle, terminando la partida.
#Si todavía quedan barcos son hundir, será el turno de la máquina. En este caso, tenemos un parámetro más: disparos_previos. Los otros dos parámetros serán el tablero_jugador y barcos_jugador, porque la máquina disparará a mi tablero.
#Si fuese nuestro turno y diaparamos a una casilla donde ya hemos disparado antes, hemos visto en la función disparar que nos dirá que ya hemos disparado ahí antes y podremos intentarlo otra vez.
#En el caso del disparo de la máquina, como lo hace "oculto" y random, mediante este parámetro disparos_previos aseguramos que el disparo de la máquina sea válido (lo hemos visto arriba, en la función turno_maquina).
#Después del turno de la máquina se hará la comprobación con la función todos_hundidos (comprobará la tabla_jugador). Si queda algún barco por hundir, el turno volverá a ser del jugador. Si todos los barcos están hundidos (devuelve True), con el break corta el bucle y la partida se acaba, siendo la máquina la ganadora.
#Resumiendo, una vez ejecutamos "jugar()", tendremos bucle hasta que un jugador gane.

def jugar():
    tablero_jugador = crear_tablero()
    tablero_maquina = crear_tablero()
    esloras = [2, 2, 2, 3, 3, 4]
    barcos_jugador = generar_barcos_aleatorios(tablero_jugador, esloras)
    barcos_maquina = generar_barcos_aleatorios(tablero_maquina, esloras)
    disparos_previos = set()
    while True:
        print("Tu tablero")
        mostrar_tablero(tablero_jugador)
        print("--------------------------")
        print("Tablero máquina")
        mostrar_tablero(tablero_maquina, ocultar_barcos=True)
        print("--------------------------")
        print("Tu turno:")
        turno_jugador(tablero_maquina, barcos_maquina)
        if todos_hundidos(tablero_maquina):
            print("¡Has ganado! ¡Has hundido todos los barcos de tu oponente!")
            break
        print("--------------------------")
        print("Turno de la máquina")
        turno_maquina(tablero_jugador, barcos_jugador, disparos_previos)
        if todos_hundidos(tablero_jugador):
            print("¡Has perdido! ¡Tu oponente ha hundido todos tus barcos!")
            break

#Ejecutamos la función jugar al darle al run.
jugar()
```
